// PERBANDINGAN DARI FURNISHING DAN STATUS MENGGUNAKAN GRAFIK BATANG
    
   Furnished,Unfurnished, dan Semi-Furnished jumlahnya sama pada status rumah Almost_ready

```python
x = list(df.iloc[:,3])
y = list(df.iloc[:,7])
plt.bar(x, y, color='blue')
plt.title("Furnished House")
plt.xlabel("Jenis Furnishing")
plt.ylabel("Status apartment")
```

// PERBANDINGAN FURNISHING DAN PRICE MENGGUNAKAN SCATTERPLOT
    
    Harga rumah Semi-Furnished mulai dari 0.1 - 2.4 milliar
    Harga rumah Furnished mulai dari 0.1 - 1.6 milliar
    Harga rumah Unfurnished mulai dari  0.1 - 1.8 milliar

```python
sns.scatterplot(data = df, x = "Furnishing", y = "Price")
plt.title("Furnished House")
plt.xlabel("Jenis")
plt.ylabel("Harga")
```

// PERSENTASI DARI FURNISHING
   
   PERSENTASI TERTINGGI DARI FURNISH RUMAH ADALAH SEMI-FURNISH, LALU UNFURNISHED DAN FURNISHED

```python
df = df.dropna(subset=['Furnishing'])
```

```python
data_pie = df['Furnishing'].value_counts().rename_axis('Furnishing').reset_index(name='Id_count')
```

```python
plt.figure(figsize =(10,10))
plt.pie(data_pie.Id_count, labels=data_pie.Furnishing, startangle=90,autopct='%.1f%%')
plt.title('Furnishing dari rumah')
plt.xlabel("Persentasi")
plt.ylabel("Persentasi")
plt.show()
```
